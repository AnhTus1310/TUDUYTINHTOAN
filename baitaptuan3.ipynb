W3A1 :Viết một chương trình nhập vào một số nguyên n và in ra các chữ số đảo ngược nhau. 

```python
n = int(input())
if n < 0:
    print("-" + str(abs(n))[::-1])
else:
    print(str(n)[::-1])
```

W3A3
Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không. 

```python
n = int(input())
print(n&(n-1)==0)
```

W3A5 viêt chương trình chia một số m cho n, kết quả thu được làm tròn lên 

```python
m,n = map(int,input().split())
a = m/n
if a == int(a):
    print(int(a))
else:
    print(int(a+1))
```

W3A6
[Even and Odd]
Nhập vào một số nguyên dương x, bạn hãy kiểm tra xem x là số chẵn hay lẻ. Nếu x là số chẵn, in ra "Even". Nếu x
là số lẻ, in ra "Odd".

```python
x = int(input())
if x%2==0:
    print("Even")
else:
    print("Odd")
```

W3A7
[Check negative numbers]
Nhập vào hai số nguyên a, b. Nếu cả hai số a, b đều âm thì in ra "Yes", ngược lại in ra "No".

```python
a,b = map(int,input().split())
if a<0 and b<0:
    print("Yes")
else:
    print("No")
```
