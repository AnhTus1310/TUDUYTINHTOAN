### Bài 1: Tìm số lớn nhất của 1 list

```python
import random
lst =[random.randrange(1,1000) for _ in range(10)]
max = -1000000
min = 1000000
for i in lst:
    if i > max:
        max =i
    if i<min:
        min = i
print(f"List da cho la:{lst}")
print(max,min)
```

### Bài 3: Tìm giá trị trung bình của 1 list

```python
import random
lst =[random.randrange(1,1000) for _ in range(10)]
tong = 0
for i in lst:
    tong +=i
sophantu = len(lst)
trungbinh = tong/sophantu
print(f"Phan tu cua chuoi la {lst}")
print(f"so trung binh la {trungbinh}")
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
lst = list(map(int,input().split()))
so0 = []
sokhac = []
for i in lst:
    if i ==0:
        so0.append(i)
    else:
        sokhac.append(i)
lst = so0 + sokhac
print(lst)
```

### Bài 6: Xóa tất cả các số giống nhau trong list, Các số khác giữ nguyên vị trí trước sau

```python
lst = list(map(int,input().split()))
res = []
for i in lst:
    if i not in res:
        res.append(i)
print(res)
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
lst = list(map(int,input().split()))
tang = True
giam = True
for i in range(len(lst)-1):
    if lst[i]>=lst[i+1]:
        tang = False
    if lst[i+1]<=lst[i]:
        giam = False
if tang:
    print("tang")
elif giam:
    print("giam")
else:
    print("vo danh")
```
